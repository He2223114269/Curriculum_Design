```python
from pyecharts.charts import Scatter
import pyecharts.options as opts
import pandas as pd

# Assuming "data" is a pandas DataFrame with columns "year", "digital", and "RD"
data = pd.read_excel("湖南省数字普惠金融与企业创新指数.xlsx")

scatter = (
    Scatter()
    .add_xaxis(data["year"].tolist())
    .add_yaxis(
        series_name="digital",
        y_axis=data["digital"],
        label_opts=opts.LabelOpts(is_show=False),
        symbol_size=10,
    )
    .add_yaxis(
        series_name="RD",
        y_axis=data["RD"],
        yaxis_index=1,
        label_opts=opts.LabelOpts(is_show=False),
        symbol_size=10,
    )
    .extend_axis(
        yaxis=opts.AxisOpts(
            name="digital",
            type_="value",
#             min_=400,
#             max_=500,
            position="left",
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(color="red")
            ),
            axislabel_opts=opts.LabelOpts(formatter="{value}", color="red"),
        )
    )
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            type_="value", 
            min_=2018,
            max_=2021,
        ),
        yaxis_opts=opts.AxisOpts(
            name="RD",
            type_="value",
#             min_=0,
#             max_=12,
            position="right",
            axisline_opts=opts.AxisLineOpts(
                linestyle_opts=opts.LineStyleOpts(color="green")
            ),
            axislabel_opts=opts.LabelOpts(formatter="{value}", color="green"),
        ),
        title_opts=opts.TitleOpts(title="Dual Axis Scatter Plot"),
    )
)

scatter.render("dual_axis_scatter.html")

scatter.render_notebook()
```
